In this notebook, we have created a 2D-circle colliding animation.
In first code block, we will import all the required imports first.

```python
from ipywidgets import HTML
import itertools
import math
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import random
```

Now, we will create a Circle class to define radius, color, speed and position of circle. This class has a motion function which can move the circle upon calling by assigning the new values to centre. Collision detection function uses the distance between the centre of two circles two find if the two circles are colliding with each other and return the boolean value.

```python
class Circle:
    ''' This class represents a circle.'''
    def __init__(self, radius: float, color: str, velocity=(0,0), centre=(0, 0)):
        self.centre = centre
        self.radius = radius
        self.velocity = velocity
        self.color = color

    def motion(self, move_tuple=(0, 0)):
        ''' It moves the circle by provided distance as tuple of points on graph or by its velocity if distance is not provided.'''
        if move_tuple == (0, 0):
            new_centre = (self.centre[0] + self.velocity[0], self.centre[1] + self.velocity[1])
        else:
            new_centre = (self.centre[0] + move_tuple[0], self.centre[1] + move_tuple[1])
        self.centre = new_centre    # position changed

    def collision_detection(self, other_circle):
        ''' This function detects if the two functions are colliding with each other and returns the boolean value.'''
        distance_bw_circles = math.sqrt(pow((self.centre[0] - other_circle.centre[0]), 2) +
                                       pow((self.centre[1] - other_circle.centre[1]), 2))
        return distance_bw_circles < (self.radius + other_circle.radius)

```

This code block will have a Simulation class which handles all the circle objects, automatically generated by this class upon providing number of circles. This class's instance will make given number of Circle objects and manipulate them using collision function. Collision function decides what will happen to the circles after they collide with each other.

```python
class Simulation:
    ''' Manages all the circle objects'''

    def __init__(self, n_circles):
        self.circles = [0]*n_circles
        self.n_circles = n_circles
        for i in range(self.n_circles):
            rad = 20
            cent = (random.randint(200, 800), random.randint(300, 600))
            vel = (random.randint(10,20), random.randint(10,20))
            colors = ['red', 'blue', 'green', 'black', 'pink']
            col = colors[random.randint(0,4)]
            self.circles[i] = (Circle(rad, col, vel, cent))

    def collision(self):
        ''' handles collision and decide what will happen in the condition.'''
        for circle in self.circles:
            circle.motion()
        for circle1, circle2 in itertools.combinations(self.circles, 2):  #idea of itertools from stack overflow
            if circle1.collision_detection(circle2) :
                circle1.velocity = (-circle1.velocity[0], circle1.velocity[1]) or (circle1.velocity[0], -circle1.velocity[1])
        for circle in self.circles:
           if circle.centre[0]>980+circle.radius or circle.centre[1]>980+circle.radius:
                circle.velocity = (0,0)
                circle.velocity =  (random.randint(-20,10), random.randint(-20,10))
           if circle.centre[0]<0+circle.radius or circle.centre[1]<0+circle.radius:
                circle.velocity = (0,0)
                circle.velocity =  (random.randint(-10,20),random.randint(-10,20))
```

Now, collide_circle function will animate the collision using funcAnimation

```python
def collide_circles(number_of_circles):
  ''' Function to animate the circles'''
  model = Simulation(number_of_circles)
  fig, ax = plt.subplots()
  ax.set_xlim(0, 1000)
  ax.set_ylim(0, 1000)
  circles_plot = [plt.Circle(circle.centre, circle.radius, color=circle.color) for circle in model.circles ]

  def update(frame):
              '''FunAnimation will change the animation according to this function.'''
              model.collision()
              for j in range(len(circles_plot)):
                  circles_plot[j].center = model.circles[j].centre
              for circle_patch in ax.patches:
                  circle_patch.remove()
              for circle_plot in circles_plot:
                  ax.add_patch(circle_plot)
              return ax.patches

  anim = FuncAnimation(fig, update, frames=100, interval=100, blit=True)
  plt.close()

  return HTML(anim.to_jshtml()) # idea from Sachjeevan
```

In this code block, we will use our code in an example.

```python
collide_circles(15)
```
